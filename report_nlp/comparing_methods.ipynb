```python
#comparing the results of two different methods

few_shot_csv = 'classified_output_AEP.csv'
bert_csv = 'BERT_results.csv'
rag_csv = 'company_results/rag_results_AEP.csv'

import pandas as pd
# Load the CSV files
few_shot_df = pd.read_csv(few_shot_csv)
bert_df = pd.read_csv(bert_csv)
rag_df = pd.read_csv(rag_csv)
```

```python
#strip empty rows from the dataframes
few_shot_df = few_shot_df.dropna(how='all')
bert_df = bert_df.dropna(how='all')
rag_df = rag_df.dropna(how='all')
```

```python
bert_df = bert_df[bert_df['company'] == 'AEP']
```

```python
rag_df_1 = rag_df[rag_df['climate_litigation'] == 'climate_litigation: 1']
rag_df_0 = rag_df[rag_df['climate_litigation'] == 'climate_litigation: 0']
```

```python
#see where there is matching text between rag_df and few_shot_df and merge the

def find_matching_text(df1, df2):
    '''note: pass the rag_df first because the few_shot is always a positive case'''
    matching_rows = []
    for index, row in df1.iterrows():
        if row['text'] in df2['text'].values:
            matching_rows.append(row)
    return pd.DataFrame(matching_rows)

matching_rag_fs = find_matching_text(rag_df, few_shot_df)
```
